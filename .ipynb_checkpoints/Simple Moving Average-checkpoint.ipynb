## Simple Moving Average

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Binance_ETHUSDT_1h.csv", index_col="date", parse_dates=True).drop(columns=["unix"])
data.head()
```

```python
data["close"].plot(figsize=(20,8))
```

```python
data["close"].iloc[::-1].plot(figsize=(20,8))
# Daily Average
data["close"].iloc[::-1].rolling(24).mean().plot(figsize=(20,8))
# Monthly Average
data["close"].iloc[::-1].rolling(24*30).mean().plot(figsize=(20,8))
```

## Covariance between the Closing Value and Volume

$$ Cov(X,Y) = E[(X - E[X])(Y - E[Y])] $$

```python
close_vol = data[["close", "Volume ETH", "Volume USDT"]].iloc[::-1].copy()
close_vol.head()
```

```python
# Taking the covariance using daily SMAs (not easy to interpret)
close_vol.rolling(24).cov().tail(12)
```

```python
close_vol["closePctChange"] = data["close"].pct_change(1)
close_vol["VolumeETHPctChange"] = close_vol["Volume ETH"].pct_change(1)
close_vol["VolumeUSDTPctChange"] = close_vol["Volume USDT"].pct_change(1)
close_vol.tail()
```

```python
close_vol[["closePctChange", "VolumeETHPctChange", "VolumeUSDTPctChange"]].rolling(24).cov().tail(12)
```
