```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("Alcohol_Sales.csv", index_col="DATE", parse_dates=True)
```

```python
data.columns = ["Sales"]
data.index.freq = 'MS'
data.head()
```

```python
N_test = 12
X_train = data["Sales"].iloc[:-N_test]
X_test = data["Sales"].iloc[-N_test:]
```

## Naive Forecast

```python
predictions = pd.DataFrame(data["Sales"].iloc[-N_test:])
predictions.columns = ["True_val"]
```

```python
naiveForecast = np.full((N_test), X_train[-1])
predictions["NV"] = naiveForecast
```

## Simple Exponential Smoothing

```python
from statsmodels.tsa.holtwinters import SimpleExpSmoothing
```

```python
ses = SimpleExpSmoothing(X_train)
```

```python
X_train.index.freq = 'MS'
ses = SimpleExpSmoothing(X_train, initialization_method='legacy-heuristic')
```

```python
fit = ses.fit(smoothing_level=0.5, optimized=False)
```

```python
predictions['SES'] = fit.predict(start=X_test.index[0], end=X_test.index[-1])
```

## Holt-Winters Model

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing
```

```python
es = ExponentialSmoothing(X_train, initialization_method='legacy-heuristic', 
                          trend='add', 
                          seasonal='add', 
                          seasonal_periods=12)
fit_es = es.fit()
```

```python
predictions['HW'] = fit_es.predict(start=X_test.index[0], end=X_test.index[-1])
```

## Comparison

```python
predictions.plot(figsize=(20,8))
```

```python
from sklearn.metrics import r2_score, mean_squared_error
```

```python
# Naive Forecast
print(r2_score(predictions['True_val'], predictions['NV']))
print(mean_squared_error(predictions['True_val'], predictions['NV'], squared=False))
```

```python
# Simple Exponential Smoothing
print(r2_score(predictions['True_val'], predictions['SES']))
print(mean_squared_error(predictions['True_val'], predictions['SES'], squared=False))
```

```python
# Holt-Winters Model
print(r2_score(predictions['True_val'], predictions['HW']))
print(mean_squared_error(predictions['True_val'], predictions['HW'], squared=False))
```
