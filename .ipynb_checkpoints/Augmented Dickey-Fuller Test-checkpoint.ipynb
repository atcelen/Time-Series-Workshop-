```python
import pandas as pd
import numpy as np
from statsmodels.tsa.stattools import adfuller
```

```python
data = pd.read_csv("Alcohol_Sales.csv", index_col="DATE", parse_dates=True)
```

```python
data.columns = ["Sales"]
data.index.freq = 'MS'
data.head()
```

```python
data["Sales"].plot(figsize=(20,8))
```

```python
adfuller(data["Sales"])
```

```python
# Try the standard normal distribution
adfuller(np.random.randn(100))[1]
```

```python
# Try a Gaussian distribution with mean=2 and standard deviation=3
adfuller(3*np.random.randn(100)+2)[1]
```

```python
# Try a Gaussian distribution with increasing variance
adfuller(np.multiply([i/10 for i in range(100)], np.random.rand(100)) + 2)[1]
```

Try logged sales data

```python
adfuller(np.log(data["Sales"]))[1]
```

Try differencing

```python
adfuller(data["Sales"].diff().dropna())[1]
```

```python
adfuller(np.log(data["Sales"]).diff().dropna())[1]
```
