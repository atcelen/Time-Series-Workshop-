```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn import linear_model
from sklearn.gaussian_process.kernels import RBF
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Create a dataset
N = 100
X = np.random.random((N, 2)) * 6 - 3 # uniformly distributed between (-3, +3)
Y = np.sin(2*X[:,0]) + np.cos(3*X[:,1])
```

$$ y = \sin(2x_1) + \cos(3x_2) $$

```python
fig = plt.figure(figsize=(20, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:,0], X[:,1], Y);
```

```python
lr = LinearRegression()
lr.fit(X,Y)
lr.coef_
```

```python
fig = plt.figure(figsize=(20, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:,0], X[:,1], Y)

# Prediction Surface
line = np.linspace(-3, 3, 50)
xx, yy = np.meshgrid(line, line)
Xgrid = np.vstack((xx.flatten(), yy.flatten())).T
Yhat = lr.predict(Xgrid).flatten()
ax.plot_trisurf(Xgrid[:,0], Xgrid[:,1], Yhat, linewidth=0.2, antialiased=True);
```

```python
ridge = linear_model.Ridge(alpha=.5)
ridge.fit(X, Y)
ridge.coef_
```

```python
fig = plt.figure(figsize=(20, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:,0], X[:,1], Y)

# Prediction Surface
line = np.linspace(-3, 3, 50)
xx, yy = np.meshgrid(line, line)
Xgrid = np.vstack((xx.flatten(), yy.flatten())).T
Yhat = ridge.predict(Xgrid).flatten()
ax.plot_trisurf(Xgrid[:,0], Xgrid[:,1], Yhat, linewidth=0.2, antialiased=True);
```

```python
gpr = GaussianProcessRegressor(kernel = RBF(1, length_scale_bounds="fixed"))
gpr.fit(X,Y)
```

```python
fig = plt.figure(figsize=(20, 8))
#ax = Axes3D(fig)
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:,0], X[:,1], Y)

# Prediction Surface
line = np.linspace(-3, 3, 50)
xx, yy = np.meshgrid(line, line)
Xgrid = np.vstack((xx.flatten(), yy.flatten())).T
mean_prediction,std_prediction = gpr.predict(Xgrid, return_std=True)
ax.plot_trisurf(Xgrid[:,0], Xgrid[:,1], mean_prediction, linewidth=0.1, antialiased=True).set_alpha(0.6)
ax.plot_trisurf(Xgrid[:,0], Xgrid[:,1], mean_prediction + 1.96 * mean_prediction, linewidth=0.1, antialiased=True).set_alpha(0.4)
#ax.plot_trisurf(Xgrid[:,0], Xgrid[:,1], mean_prediction - 1.96 * mean_prediction, linewidth=0.1, antialiased=True).set_alpha(0.2)
```
