```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import acf
from statsmodels.graphics.tsaplots import plot_pacf
```

```python
x0 = np.random.randn(1000)
```

```python
plt.plot(x0)
```

```python
plot_pacf(x0, method='ywm')
```

```python
# Try a function with a time dependency
x1 = [0]
for i in range(1000):
  x = 0.5 * x1[-1] + 0.1 * np.random.randn()
  x1.append(x)
x1 = np.array(x1)
```

```python
plt.plot(x1)
```

```python
plot_pacf(x1)
```

```python
# Try a negative correlation
x1 = [0]
for i in range(1000):
  x = -0.5 * x1[-1] + 0.1 * np.random.randn()
  x1.append(x)
x1 = np.array(x1)
```

```python
plt.plot(x1)
```

```python
plot_pacf(x1)
```

```python
# Try a correlation from far points
x1 = list(np.random.randn(10))
for i in range(1000):
  x = -0.5 * x1[-7] + 0.3 * x1[-5] + 0.1 * np.random.randn()
  x1.append(x)
x1 = np.array(x1)
```

```python
plt.plot(x1)
```

```python
plot_pacf(x1)
```
